```python
# Task:
# create a class named as Atm
# add the following data/Attributes -> account_balance, atm_pin
# add the following behaviour/methods -> deposite, withdraw, balance

# attributes are fucntions inside a class

class Atm:
    def __init__(self):
        self.pin = int()
        self.balance = 0
        self.menu()

    def menu(self):

        user_input = input('''
                    Hello How would you like to procees. Please pick from the following options
                           1. Create Pin \n
                           2. Deposite \n
                           3. Withdraw \n
                           4. Check balance \n
                           5. exit \n
        ''')
    
        if user_input == '1':
            self.create_pin()
        elif user_input == '2':
            self.deposit()
        elif user_input == '3':
            self.withdraw()
        elif user_input == '4':
            self.check_balance()
        else:
            print('Have a good day!!')
    
    def create_pin(self):
        self.pin = int(input('Enter New Pin:'))
        print('Pin set successfully')

    def deposit(self):
        temp = int(input('Enter your pin'))
        if temp == self.pin:
            amount = int(input('Enter amount to deposit'))
            self.balance += amount
        else:
            print('Invalid PIN')
    
    def withdraw(self):
        temp = int(input('Enter your pin'))
        if temp == self.pin:
            amount = int(input('Enter the amount to withdraw:'))
            if amount <= self.balance:
                self.balance -= amount
            else:
                print('Insufficient balance')
        else: 
            print('Incorrect PIN')
    
    def check_balance():
        temp = int(input('Enter your pin:'))
        if temp == self.pin:
            print(self.balance)
        else:
            print('Invalid PIN')
    
```
