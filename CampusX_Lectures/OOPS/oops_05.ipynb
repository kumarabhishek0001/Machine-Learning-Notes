### Inheritance

- a children class can inherit Data members, methods, constructors from the parent.
- private members (methods or variables) are not inherited.

```python
class User:

    
    def login(self):
        print('logged in')

    def register(self):
        print('registred!!')

class Student(User):
    def enroll(self):
        print('enrolled')
    def review(self):
        print('review given!!')

stu1 = Student()
```

```python
# methods inherited from User class
stu1.login()
stu1.register()

# methods inside student class
stu1.enroll()
stu1.review()
```
