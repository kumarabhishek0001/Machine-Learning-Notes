```python
class Fraction:
    def __init__(self,n,d):
        self.num = n
        self.den = d
    # just after this line try to print this by making object and telling to print

    def __str__(self):
        return f"{self.num}/{self.den}"
    
    def __add__(self,other):
        temp_num = self.num*other.den + self.den*other.num
        temp_den = self.den*other.den

        return f"{temp_num}/{temp_den}"

```

```python
frac1 = Fraction(1,2)
print(frac1)

# python follows lazy loading -> resources are not called until neccessary. here though this happens because we haven't told the function what to do when print is called
# so to tell fucntion what to do when print is called we use __str__
```

```python
frac2 = Fraction(2,3)
print(frac2)

frac3 = Fraction(4,5)
print(frac3)
```

```python
print(frac2+frac3)
# this does not run since we haven't defined behaviour for addition-> for this we use __add__
```
