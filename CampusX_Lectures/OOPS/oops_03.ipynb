```python
# creating customer class
class Customer:
    def __init__(self,name):
        self.name = name

# creating object of class Customer
cust = Customer('Abhishek')
# print(cust.name)

# creating greet function
def greet(customer):
    print(customer.name)

# passing reference cust
# this means we can pass object as a parameter
greet(cust)

# cust(reference variable) -----> Object in memory <------- customer(parameter of greet, pointing to object/instance)
# any changes that comes eitht from cust or customer(parameter of greet) will cause change of object here will be reflected in both no matter what
```

```python
class Customer:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender

def greet(customer):
    if customer.gender == 'male':
        print('hello', customer.name, 'sir')
    else:
        print('hello', customer.name, 'madam')

cust = Customer('ankita', 'female')
greet(cust)
```

```python
# we can also return an object
def greet(customer):
    if customer.gender == 'male':
        print('hello', customer.name, 'sir')
    else:
        print('hello', customer.name, 'madam')

    cust2 = Customer('Ankit', 'male')
    return cust2

cust = Customer('ankita', 'female')
new_cust = greet(cust)

print("new customer: ",new_cust.name, new_cust.gender)
```

```python
# creating class called customer
class Customer:
    def __init__(self,name):
        self.name = name

def greet(customer):
    print(f"hello {customer.name}")
    print(f"id of customer inside greet function: {id(customer)}")


# creating object and referencing it
cust = Customer('Ankit')
print(f"cust referencing to object id: {id(cust)}")

greet(cust)

# Both cust (outside) and customer (inside function) refer to the same instance of Customer class
# Therefore, id(cust) == id(customer)
```

```python
# chaniging the attributes when reference is passed -> MUTATION VIA REFERENCE   

class Customer:
    def __init__(self,name):
        self.name = name

def greet(customer):
    print(f"hello {customer.name}")
    print(f"id of customer inside greet function: {id(customer)}")


    customer.name = 'Ankita'
    print(f"hello {customer.name}")
    print(f"id of customer inside greet function: {id(customer)}")

cust = Customer('Ankit')
print(f"cust referencing to object id: {id(cust)}")


greet(cust.name)

# when referce is passed changes takes place in original data that is ankit
```

```python
# collection of objects

class Customer:
    def __init__(self,name, gender):
        self.name = name
        self.gender = gender

c1 = Customer('Nitish', 34)
c2 = Customer('Ankit', 23)
c3 = Customer('Ankita', 30)

customer_list = [c1, c2, c3]

for i in customer_list:
    print(i.name, i.gender)
```
