```python
# THERE MAY COME SCENARIO WHERE THE DEVELOPER HAS TO ACCESS IT.
# we use GETTER and SETTER methods for that.
# but why to give access now. 
# we can give access now because with these to we can ADD LOGICS AND SAFETY CHECKS. before any modification

# Task:
# create a class named as Atm
# add the following data/Attributes -> account_balance, atm_pin
# add the following behaviour/methods -> deposite, withdraw, balance

# attributes are fucntions inside a class

class Atm:
    def __init__(self):
        self.__pin = int()
        self.__balance = 0
        self.menu()

    def menu(self):

        user_input = input('''
                    Hello How would you like to procees. Please pick from the following options
                           1. Create Pin \n
                           2. Deposite \n
                           3. Withdraw \n
                           4. Check balance \n
                           5. exit \n
        ''')
    
        if user_input == '1':
            self.create_pin()
        elif user_input == '2':
            self.deposit()
        elif user_input == '3':
            self.withdraw()
        elif user_input == '4':
            self.check_balance()
        else:
            print('Have a good day!!')
    
    def get_pin(self):
        return self.__pin

    def set_pin(self, new_pin):
            if type(new_pin) == int:
                self.__pin = new_pin
            else:
                print('New PIN is invalid')
    
    def create_pin(self):
        self.__pin = int(input('Enter New Pin:'))
        print('Pin set successfully')

    def deposit(self):
        temp = int(input('Enter your pin'))
        if temp == self.__pin:
            amount = int(input('Enter amount to deposit'))
            self.__balance += amount
        else:
            print('Invalid PIN')
    
    def withdraw(self):
        temp = int(input('Enter your pin'))
        if temp == self.__pin:
            amount = int(input('Enter the amount to withdraw:'))
            if amount <= self.__balance:
                self.__balance -= amount
            else:
                print('Insufficient balance')
        else: 
            print('Incorrect PIN')
    
    def check_balance(self):
        temp = int(input('Enter your pin:'))
        if temp == self.__pin:
            print(self.__balance)
        else:
            print('Invalid PIN')
    
```
