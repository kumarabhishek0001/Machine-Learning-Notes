## Static Method
- make a counter
- whenever a new object is created this counter auto increments
- this counter is used to make serial numbers for different account

```python
class Atm:
    def __init__(self):
        self.__pin = ''
        self.__balance = 0

        self.sno = 1
        self.sno += 1

c1 = Atm()
print(c1.sno )

c2 = Atm()
print(c2.sno)

c3 = Atm()
print(c3.sno)
```

```python
# to avoid the problem above we make static variables
# -> instance variables are those variables which have differen value for each instancce
# -> static variables are those variables whose value remains same accross multiple instances

class Atm:
    # static varaiable
    counter = 1


    def __init__(self):
        # Instance Variable
        self.__pin = ''
        self.__balance = 0

        self.sno = Atm.counter
        Atm.counter += 1

c1 = Atm()
print(c1.sno )

c2 = Atm()
print(c2.sno)

c3 = Atm()
print(c3.sno)
```

```python
class Atm:
    # static varaiable
    __counter = 1


    def __init__(self):
        # Instance Variable
        self.__pin = ''
        self.__balance = 0

        self.sno = Atm.__counter
        Atm.__counter += 1
    
    @staticmethod
    def get_counter():
        return Atm.__counter 
    
    @staticmethod
    def set_counter(new_counter):
        if type(new_counter) == int:
            Atm.__counter = new_counter
        else: 
            print('invalid value entered')
```
